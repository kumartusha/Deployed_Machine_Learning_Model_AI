# Feature Engineering for the dataset.

```python
# Do Some Feature Engineering for the dataset.
dataset["Year"] = dataset["Date"].dt.year
dataset["Month"] = dataset["Date"].dt.month
dataset["Day"] = dataset["Date"].dt.day

#  Drop the Date Column.
dataset.drop("Date", axis=1, inplace=True)
```

```python
#  Do some advance feature Engineering for the dataset.
dataset["Temp_diff"] = dataset["MaxTemp"] - dataset["MinTemp"]
dataset["Temp_3pm_9am_diff"] = dataset["Temp3pm"] - dataset["Temp9am"]
dataset["Cloud_avg"] = (dataset["Cloud3pm"] + dataset["Cloud9am"]) / 2
dataset["Humidity3pm_9am_Diff"] = dataset["Humidity3pm"] - dataset["Humidity9am"]
dataset["Pressure3pm_9am_Diff"] = dataset["Pressure3pm"] - dataset["Pressure9am"]
```

```python
dataset.head()
```

```python
#  Now Remove all the Featured Columns from the dataset.
dataset.drop(columns=["MinTemp", "MaxTemp", "Temp3pm", "Temp9am", "Cloud3pm", "Cloud9am", "Humidity3pm", "Humidity9am", "Pressure3pm", "Pressure9am"], axis=1, inplace=True)
```

```python
#  Divide the Features of the dataset into the numerical and the categorical 
categorical_col = list(dataset.select_dtypes(include="object").columns)
numerical_col = [i for i in dataset.columns if i not in categorical_col]
```

## Data Preprocessing.

```python
dataset[numerical_col].isnull().sum()
```

```python
#  Fill the mean in all the numeric columns.

for i,val in enumerate(numerical_col):
    dataset[val].fillna(dataset[val].mean(), inplace=True)
```

```python
dataset[numerical_col].isnull().sum()
```

```python
#  Fill the mean in all the categorical columns.

for i,val in enumerate(categorical_col):
    dataset[val].fillna(dataset[val].mode()[0], inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.head()
```

```python
#  Perform the Label Encoding for the Categorical Features.

encoder = LabelEncoder()

for i,val in enumerate(categorical_col):
    dataset[val] = encoder.fit_transform(dataset[val])

#  Now our data will be encoded.
```

```python
dataset.head()
```

```python
dataset.duplicated().sum()
```

```python
dataset.drop_duplicates(inplace=True)
```

```python
dataset.head()
```

### Splitting into the dependent and the independent Variable.

```python
# Splitting into the dependent and the independent Variable.
x = dataset.drop("RainTomorrow", axis=1)
y = dataset["RainTomorrow"]
```

```python
print(x.shape, y.shape)

print(y.value_counts())
```

## Split the dataset into the training and the testing Dataset.

```python
y.value_counts()
```

```python
#  Apply the smote first on the dependent and the inependent Variable.
#  Apply the Smote Over Sampling Technique for Handling the data imbalance.

smote = SMOTE(sampling_strategy="auto", random_state=42)

x_smoted , y_smoted = smote.fit_resample(x, y)

print(x.shape, y.shape)
print(y_smoted.value_counts())
len(y_smoted)
```

```python
x_train_smoted, x_test_smoted, y_train_smoted, y_test_smoted = train_test_split(x_smoted, y_smoted, test_size=0.2, random_state=42)
```

```python
print(x.shape, x_train_smoted.shape, y_train_smoted.shape)
print(x.shape, x_test_smoted.shape, y_test_smoted.shape)
```

## Model Implementation and Training

#### Logistic Regression Model Training

```python
model1 = LogisticRegression()

model1.fit(x_train_smoted, y_train_smoted)
```

```python
# Find the accuracy for the Model.

print("The accuracy for the training data are:- ",accuracy_score(model1.predict(x_train_smoted), y_train_smoted))
model1Acc = accuracy_score(model1.predict(x_test_smoted), y_test_smoted)
print("The accuracy for the testing data are:- ",accuracy_score(model1.predict(x_test_smoted), y_test_smoted))
print("The confusion matrix prediction are:", confusion_matrix(model1.predict(x_test_smoted), y_test_smoted))
print("The classification report for model are:", classification_report(model1.predict(x_test_smoted), y_test_smoted))
```

### XGB Classifier Algo

```python
model6 = XGBClassifier()
```

```python
parameters = {
   "n_estimators": [50, 100, 200],  # Number of boosting rounds
    "learning_rate": [0.01, 0.1, 0.2],  # Step size shrinkage
    "max_depth": [3, 6, 9],  # Maximum depth of trees
    "min_child_weight": [1, 3, 5],  # Minimum sum of instance weights in a child
    "subsample": [0.6, 0.8, 1.0],  # Fraction of samples for training
    "colsample_bytree": [0.6, 0.8, 1.0],  # Fraction of features per tree
    "gamma": [0, 0.1, 0.2],  # Minimum loss reduction for a split
}
```

```python
#  Apply the Hyperparameter Tuning.
random_search = RandomizedSearchCV(model6, parameters, cv=5, n_jobs=-1, verbose=2)
random_search.fit(x_train_smoted, y_train_smoted)
```

```python
random_search.best_params_
```

```python
random_search.best_score_
```

```python
model6 = XGBClassifier(subsample = 1, n_estimators = 200, min_child_weight = 5, max_depth = 9, learning_rate = 0.1, gamma = 0.1, colsample_bytree = 0.8)
model6.fit(x_train_smoted, y_train_smoted)
```

```python
# # Find the accuracy for the Model.

print("The accuracy for the training data are:- ",accuracy_score(model6.predict(x_train_smoted), y_train_smoted))
model6Acc = accuracy_score(model6.predict(x_test_smoted), y_test_smoted)
print("The accuracy for the testing data are:- ",accuracy_score(model6.predict(x_test_smoted), y_test_smoted))
print("The confusion matrix prediction are:", confusion_matrix(model6.predict(x_test_smoted), y_test_smoted))
print("The classification report for model are:", classification_report(model6.predict(x_test_smoted), y_test_smoted))
```
